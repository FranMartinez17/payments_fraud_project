The results obtained are very similar for the F1 Score. 

For the ROC_AUC all the models with the exception of the base and the lightGBM are above 0.95.

If we take into account the false positives ratio, the lightGBM underperformed with a 56%. This percentage differs a lot from the others models as XGBoost, AdaBoost and Random Forest are belos the 10%.

For the True positives it is worth higlighting that the Random Forest obtained 1.

If we take a look at the F2 score, it is apreciated that this metric is the one which diffirienciate most the models. Random Forest obtained the best result with a 0.764, followed by the SVM and the XGBoost with 0.638 and 0.626.

After this results, the Random Forest will be selected for its optimization, as it has obtained the best results overall and its F2 score, which takes into account the false negatives, is the highest with more than 10% difference.
